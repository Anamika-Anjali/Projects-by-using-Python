# Exploratory Data Analysis

```python
df.columns
```

# Gender

```python
sns.countplot(x = 'Gender',data = df)
```

```python
ax = sns.countplot(x = 'Gender',data = df)
for bar in ax.containers:
    ax.bar_label(bar)
```

According to this graph, there are the most male customers.

```python
ax = sns.countplot(data = df ,x ='Age',hue = 'Gender' )
sns.set(rc={'figure.figsize':(15,5)})
```

According to this graph, the most customers, both male and female, are in the 36-37 age group.

# Region

```python
ax = sns.countplot(data = df, x = 'Region', hue = 'Job Classification')
sns.set(rc={'figure.figsize':(15,10)})

for bars in ax.containers:
    ax.bar_label(bars)
```

According to this graph, the most female customers are from England.

# Job Classification

```python
ax = sns.countplot(data = df, x = 'Region', hue = 'Job Classification')
sns.set(rc={'figure.figsize':(15,10)})

for bars in ax.containers:
    ax.bar_label(bars)
```

According to this graph, the most job classification is white collar, which is from England.

# Balance 

```python
creg_join = df.groupby(['Balance'],as_index=False).sum().sort_values(by='Gender',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data =creg_join, x='Balance',y='Gender')
```

According to this graph, the highest balance is for males.

# Conclusion

The data shows that England has the highest number of bank customers, particularly in the 36-37 age group. Male customers dominate this group and have higher account balances. Furthermore, most of these customers belong to the white-collar job category.
