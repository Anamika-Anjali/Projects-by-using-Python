# Exploratory Data Analysis

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df,x='region',order=df['region'].value_counts().index)
plt.title("Distribution of Dishes by region")
plt.xlabel("region")
plt.ylabel("number of dishes")
plt.show()     
```

According to this graph, the highest number of diseases are in the West.

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df,x='diet')
plt.title("Distribution of Diet types")
plt.xlabel("Diet type")
plt.ylabel("Number of dishes")
plt.show()
     
     
```

According to this graph, the highest number of diseases are in vegetarians.

```python
ax = sns.countplot(data = df, x = 'diet', hue = 'region')
sns.set(rc={'figure.figsize':(20,10)})
plt.title("Total vegeterian and non veg orders by region")

for bars in ax.containers:
    ax.bar_label(bars)
```

According to this graph, the highest number of vegetarian orders are placed by people from the West.

```python
ax = sns.countplot(x = 'flavor_profile',data = df)
plt.title("Flavor_profile")

for bar in ax.containers:
    ax.bar_label(bar)
```

According to this graph, the highest preference for spicy food is observed.
