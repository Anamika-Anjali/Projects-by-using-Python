# Exploratory Data Analysis

# Gender

```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

# Alcohol_Consumptionm

According to the given graph, the health effects on others and females are higher compared to males.

```python
ax = sns.countplot(x = 'Alcohol_Consumption',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

# Smoking_Habits

```python
ax = sns.countplot(data = df, x = 'Gender', hue = 'Smoking_Habits')
sns.set(rc={'figure.figsize':(15,10)})

for bars in ax.containers:
    ax.bar_label(bars)
```

According to the given graph, females are most affected by smoking habits.

# Health_Issue_Type

```python
ax = sns.countplot(data = df, x = 'Gender', hue = 'Health_Issue_Type')
sns.set(rc={'figure.figsize':(15,10)})

for bars in ax.containers:
    ax.bar_label(bars)
```

According to the given graph, the most affected by COPD are females, and others are affected by alcohol poisoning.



```python
ax = sns.countplot(data = df, x = 'Health_Issue_Type', hue = 'Age')
sns.set(rc={'figure.figsize':(15,5)})

for bars in ax.containers:
    ax.bar_label(bars)
```
